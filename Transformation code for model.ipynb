```python
def top_tweets(df,by,k):
    df_agg = df.groupby(['date','clean_tweet'])
    df1 = df_agg.first()
    df2 = df1[[by]]
    data_sorted = df2.sort_values(by =['date',by],ascending = False).apply(list)
    rdf = data_sorted.head(k)
    new_df = rdf.transpose()
    c =  new_df.unstack().reset_index(level = 0).reset_index()
    final = c[['date','clean_tweet']]
    final_df = final.transpose()
    final_df.reset_index(drop=True, inplace=True) 
    df = final_df.transpose()
    df.columns = ['Date','tweet']
    top_tweetsdf = df.pivot_table(index = 'Date', columns = df.groupby('Date').cumcount().add(1), 
                values = 'tweet', aggfunc='first').add_prefix('Top ').add_suffix(' Tweet').rename_axis(columns=None).reset_index()
 
    return top_tweetsdf


   
```

```python
top_tweets(df_final,by ='finance',k = 30)
```
